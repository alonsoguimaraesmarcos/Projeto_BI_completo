```python
import pandas as pd
import numpy as np
```

```python
# Loading files
df_17 = pd.read_csv('acidentes2017.csv', engine= 'python', sep = ';')
df_18 = pd.read_csv('acidentes2018.csv', engine= 'python', sep = ';')
df_19 = pd.read_csv('acidentes2019.csv', engine= 'python', sep = ';')
df_20 = pd.read_csv('acidentes2020.csv', engine= 'python', sep = ';')
```

```python
# Concat files as "df".
files_prf = [df_17,df_18,df_19,df_20]
df = pd.concat(files_prf)
```
